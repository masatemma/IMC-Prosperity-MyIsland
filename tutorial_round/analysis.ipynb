```python
import pandas as pd
df = pd.read_csv('results/example_trader_results.csv', sep=(';'))
amethysts_data = df[df['product'] == "AMETHYSTS"]
amethysts_data
```

```python
import matplotlib.pyplot as plt
import numpy as np
# define data values
amethysts_data = df[df['product'] == "AMETHYSTS"]
amethysts_data

x_sequence = range(1, len(amethysts_data) + 1)

plt.figure(figsize=(10, 6))
plt.plot(x_sequence, amethysts_data['mid_price'], label='AMETHYSTS Mid Price')
plt.ylim([9990, 10010])
```
