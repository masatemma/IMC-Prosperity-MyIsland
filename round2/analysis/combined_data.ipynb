```python
import re
import json
import pandas as pd
from sklearn.linear_model import LinearRegression
import numpy as np
import io
import csv
from collections import defaultdict
import matplotlib.pyplot as plt
```

```python
files = ['../past_data/prices_round_2_day_-1.csv', '../past_data/prices_round_2_day_0.csv', '../past_data/prices_round_2_day_1.csv']
dataframes = [pd.read_csv(file, delimiter=';') for file in files]
# Concatenate all DataFrames into one
combined_df = pd.concat(dataframes, ignore_index=True)
combined_df
combined_df['SUNLIGHT'].mean()
combined_df['ORCHIDS'].mean()
```

```python
# Create a plot with a primary y-axis
fig, ax1 = plt.subplots(figsize=(12, 6))

# Plotting the price of orchids on the primary y-axis
color = 'tab:blue'
ax1.set_xlabel('Row Number')
ax1.set_ylabel('Price of Orchids', color=color)
ax1.plot(combined_df.index, combined_df['ORCHIDS'], color=color)
ax1.tick_params(axis='y', labelcolor=color)

# Create a second y-axis for sunlight
ax2 = ax1.twinx()
color = 'tab:orange'
ax2.set_ylabel('Sunlight', color=color)
ax2.plot(combined_df.index, combined_df['SUNLIGHT'], color=color)
ax2.tick_params(axis='y', labelcolor=color)

# Create a third y-axis for humidity
ax3 = ax1.twinx()
# Offset the third axis to ensure it does not overlap with the second
ax3.spines['right'].set_position(('outward', 60))
color = 'tab:green'
ax3.set_ylabel('Humidity', color=color)
ax3.plot(combined_df.index, combined_df['HUMIDITY'], color=color)
ax3.tick_params(axis='y', labelcolor=color)

# Adding a title and a tight layout
plt.title('Price, Sunlight, and Humidity Trends Over Rows')
fig.tight_layout()  # Adjusts plot to minimize overlap and clipping

# Show plot
plt.show()
```
