```python
import re
import json
import pandas as pd
from sklearn.linear_model import LinearRegression
import numpy as np
import io
import csv
from collections import defaultdict
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
# Load the CSV file
filepath = '../past_data/prices_round_2_day_-1.csv'
df = pd.read_csv(filepath, delimiter=';')

# Filter the DataFrame to include only rows where humidity is between 60 and 80
filtered_df = df[(df['HUMIDITY'] >= 60) & (df['HUMIDITY'] <= 80)]

# Save the filtered DataFrame to a new CSV file

# Create a plot with a primary y-axis
fig, ax1 = plt.subplots(figsize=(12, 6))

# Plotting the price of orchids on the primary y-axis
color = 'tab:blue'
ax1.set_xlabel('Row Number')
ax1.set_ylabel('Price of Orchids', color=color)
ax1.plot(filtered_df.index, filtered_df['ORCHIDS'], color=color)
ax1.tick_params(axis='y', labelcolor=color)

# Create a second y-axis for sunlight
ax2 = ax1.twinx()
color = 'tab:orange'
ax2.set_ylabel('Sunlight', color=color)
ax2.plot(filtered_df.index, filtered_df['SUNLIGHT'], color=color)
ax2.tick_params(axis='y', labelcolor=color)


# Adding a title and a tight layout
plt.title('Price, Sunlight Trends Over Rows')
fig.tight_layout()  # Adjusts plot to minimize overlap and clipping

# Show plot
plt.show()
```

```python

files = ['../past_data/prices_round_2_day_-1.csv', '../past_data/prices_round_2_day_0.csv', '../past_data/prices_round_2_day_1.csv']
dataframes = [pd.read_csv(file, delimiter=';') for file in files]
# Concatenate all DataFrames into one
combined_df = pd.concat(dataframes, ignore_index=True)



# Assuming 'SUNLIGHT' and 'HUMIDITY' are your independent variables and 'ORCHIDS' is the dependent variable
X = combined_df[['SUNLIGHT', 'HUMIDITY']]
y = combined_df['ORCHIDS']

# Add a constant to the model (the intercept)
X = sm.add_constant(X)

# Create the model
model = sm.OLS(y, X).fit()

# Print the summary of the regression
print(model.summary())
```
