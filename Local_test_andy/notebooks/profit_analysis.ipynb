# Superimpose

```python
df_aggresell_1_buypeak
df_aggresell_2_buypeak
df_peaks_only
df_sellpeak_agg_2til10_buypeak
```

```python
df_superimpose = pd.DataFrame()
df_superimpose['timestamp'] = df_peaks_only['timestamp']
df_superimpose['peaks_only'] = df_peaks_only['profit_and_loss']
df_superimpose['as2_bp'] = df_aggresell_2_buypeak['profit_and_loss']
df_superimpose['as1_bp'] = df_aggresell_1_buypeak['profit_and_loss']
df_superimpose['sp_as2til10_bp'] =  df_sellpeak_agg_2til10_buypeak['profit_and_loss']
df_superimpose
```

```python
# Plot scatter graph with superimposed points
ax = df_superimpose.plot.scatter(x='timestamp', y='peaks_only', color='red', label='peaks_only', s=1)
df_superimpose.plot.scatter(x='timestamp', y='as2_bp', color='green', label='as2_bp', s=1, ax=ax)
df_superimpose.plot.scatter(x='timestamp', y='as1_bp', color='blue', label='as1_bp', s=1, ax=ax)
df_superimpose.plot.scatter(x='timestamp', y='sp_as2til10_bp', color='purple', label='sp_as2til10_bp', s=1, ax=ax)

# Add labels and title
plt.xlabel('Timestamp')
plt.ylabel('Value')
plt.title('Superimposed Scatter Plot')

# Show the plot
plt.grid(True)
plt.legend()
plt.show()
```
