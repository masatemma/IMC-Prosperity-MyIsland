```python
BID_L_list = [i for i in range(1, 100)]
BID_H_list = [i for i in range(1, 100)]
NUM_TOTAL = 1000

def cdf(x):
    y = x/5000
    return x*y/2

# Made tuples of BID_L AND BID_H
bid_tuple_list= list()
for i in range(len(BID_L_list)):
    for j in range(i, len(BID_H_list)):
        bid_tuple_list.append((BID_L_list[i], BID_H_list[j]))

profit_list = list()
for bid_tuple in bid_tuple_list:
    BID_L = bid_tuple[0]
    BID_H = bid_tuple[1]
    E_L = NUM_TOTAL * cdf(BID_L)
    E_H = NUM_TOTAL * cdf(BID_H)
    avg_cost = (BID_L * E_L + BID_H * (E_H - E_L))/E_H
    profit_list.append((1000-(avg_cost+900))*E_H)

max_idx = 0
for i in range(1, len(profit_list)):
    if profit_list[i] > profit_list[max_idx]:
        max_idx = i

print(max_idx)
print(bid_tuple_list[max_idx])
print(profit_list[max_idx])
```
