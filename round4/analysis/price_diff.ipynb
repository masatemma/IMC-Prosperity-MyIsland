```python
import re
import json
import pandas as pd
import io
import csv
from collections import defaultdict
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv("../past_data/combined_coconut_coupon_prices.csv")
data['Price_Diff'] = data['COCONUT_COUPON'] + 10000 - data['COCONUT']
df = data[3000:4000]

```

```python
import matplotlib.pyplot as plt

# Plot the 'New_Value' column
plt.figure(figsize=(10, 6))
plt.plot(df['Price_Diff'], label='Price_Diff Value')
plt.title('Price_Diff Plot')
plt.xlabel('Index')
plt.ylabel('Value')
plt.legend()
plt.grid(True)

# Display the plot
plt.show()  # To render the plot
```

```python

# Plot 'COCONUT_COUPON', 'COCONUT', and 'Price_Diff' with separate y-axis labels and vertical grid lines for each x-axis label
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plotting COCONUT on the left y-axis with vertical lines
color = 'tab:blue'
ax1.set_xlabel('Index')
ax1.set_ylabel('COCONUT', color=color)
ax1.plot(df['COCONUT'], color=color, label='COCONUT')
ax1.tick_params(axis='y', labelcolor=color)
ax1.grid(True, which='both', axis='x', color='gray', linestyle='--')  # Vertical lines on x-axis

# Creating the second y-axis for COCONUT_COUPON with vertical lines
ax2 = ax1.twinx()  # Create a second axis that shares the same x-axis
color = 'tab:orange'
ax2.set_ylabel('COCONUT_COUPON', color=color)
ax2.plot(df['COCONUT_COUPON'], color=color, label='COCONUT_COUPON')
ax2.tick_params(axis='y', labelcolor=color)

# Creating a third y-axis for Price_Diff
ax3 = ax1.twinx()  # Create a third axis that shares the same x-axis
ax3.spines['right'].set_position(('outward', 60))  # Offset the new y-axis to prevent overlapping
ax3.set_ylabel('Price_Diff', color='tab:green')
ax3.plot(df['Price_Diff'], color='tab:green', linestyle='--', label='Price_Diff')
ax3.tick_params(axis='y', labelcolor='tab:green')

# Add a title and ensure vertical lines for each x-axis label
plt.title('COCONUT_COUPON, COCONUT, and Price_Diff with Vertical Lines')
plt.grid(True, which='both', axis='x', color='gray', linestyle='--')
ax3.legend(loc='upper left')

# Display the plot with COCONUT, COCONUT_COUPON, and Price_Diff with separate y-axis labels
plt.show()  # To render the plot
```
