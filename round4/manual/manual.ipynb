```python
BID_L_list = [i for i in range(1, 100)]
BID_H_list = [i for i in range(1, 100)]
NUM_TOTAL = 100000

def cdf(x):
    y = x/5000
    return x*y/2

# Made tuples of BID_L AND BID_H
bid_list= list()
for i in range(len(BID_L_list)):    
    bid_list.append(BID_L_list[i])

profit_list = list()
for bid in bid_list:
    BID_L = bid    
    E_L = NUM_TOTAL * cdf(BID_L)    
    avg_cost = (BID_L * E_L)/E_L
    profit_list.append((1000-(avg_cost+900))*E_L)

max_idx = 0
for i in range(1, len(profit_list)):
    if profit_list[i] > profit_list[max_idx]:
        max_idx = i

print(max_idx)
print(bid_list[max_idx])
print(profit_list[max_idx])
```

```python


def cdf(x):
    y = x/5000
    return x*y/2

def best_comb(average_bid):
    BID_L_list = [i for i in range(1, 100)]
    BID_H_list = [i for i in range(1, 100)]
    NUM_TOTAL = 5000

    # Made tuples of BID_L AND BID_H
    bid_tuple_list= list()
    for i in range(len(BID_L_list)):
        for j in range(i, len(BID_H_list)):
            bid_tuple_list.append((BID_L_list[i], BID_H_list[j]))
    profit_list = list()
    for bid_tuple in bid_tuple_list:
        BID_L = bid_tuple[0]
        BID_H = bid_tuple[1]
        E_L = NUM_TOTAL * cdf(BID_L)
        E_H = NUM_TOTAL * cdf(BID_H)
        low_bid_profit = (1000 - (BID_L + 900))*E_L
        high_bid_profit = (1000 - (BID_H + 900))*(E_H - E_L)    
        if BID_H < average_bid:
            high_bid_profit = high_bid_profit * ((1000 - average_bid) / (1000 - BID_H))
            
        total_profit = low_bid_profit + high_bid_profit        
        if BID_L < BID_H:
            profit_list.append(total_profit)
        else:
            profit_list.append(-1)

    max_idx = 0
    for i in range(1, len(profit_list)):
        if profit_list[i] > profit_list[max_idx]:
            max_idx = i

    # print(max_idx)
    # print(bid_tuple_list[max_idx])
    # print(profit_list[max_idx])
    # index = bid_tuple_list.index((52,79))
    # print(profit_list[index])
    print(f"AVG second bid: {average_bid}, best_comb: {bid_tuple_list[max_idx]} best_profit: {profit_list[max_idx]}")
    


for i in range(50, 100):
    best_comb(i)
#best_comb(78)

```
